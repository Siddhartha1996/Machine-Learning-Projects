# Lets do Linear Regression:

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df1.drop('Human Development Index (HDI)',axis=1)
```

```python
X
```

```python
y = df1['Human Development Index (HDI)']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print('R-squared: ',r2_score(y_test,pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
sns.regplot(y_test,pred)
```

```python
# ElasticNet Regression
```

```python
from sklearn.linear_model import ElasticNet
clt=ElasticNet()
clt.fit(X_train,y_train)
y_pred=clt.predict(X_test)
```

```python
y_pred
```

```python
print('R-squared: ',r2_score(y_test,y_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
sns.regplot(y_test,y_pred)
```

```python
#Linear Regression Worked Much Better.
```

```python
#Ridge Regression:

from sklearn.linear_model import Ridge
model=Ridge()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
print('R-squared: ',r2_score(y_test,y_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
sns.regplot(y_test,y_pred)
```

```python
# Ridge Regression is the best so far!
```

```python
#Lasso:


from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(X_train,y_train)
lassopred=lasso.predict(X_test)
```

```python
print('R-squared: ',r2_score(y_test,lassopred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,lassopred)))
```

```python
sns.regplot(y_test,lassopred)
```
