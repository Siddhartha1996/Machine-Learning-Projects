# Data Preparation:

```python
df1 = df.drop(['Make','Model','Origin'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Type']=le.fit_transform(df1['Type'])
```

```python
df1
```

```python
df1['DriveTrain']=le.fit_transform(df1['DriveTrain'])
```

```python
df1
```

```python
df1 = df1.drop('Invoice',axis=1)
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df2 = sc.fit(df1)
df2
df1
```

```python
df1= pd.get_dummies(df1)
```

```python
y=df1['MSRP']
X=df1.drop('MSRP',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=50)
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print('R-squared: ',r2_score(y_test,pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
sns.regplot(y_test,pred)
```

```python
# Our goodness of fit is .775
```

```python
#Ridge Regression:

from sklearn.linear_model import Ridge
model=Ridge()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
print('R-squared: ',r2_score(y_test,y_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python

#Lasso:


from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(X_train,y_train)
lassopred=lasso.predict(X_test)
```

```python
print('R-squared: ',r2_score(y_test,lassopred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,lassopred)))
```

```python
sns.regplot(y_test,lassopred)
```

```python
# ElasticNet Regression
```

```python
from sklearn.linear_model import ElasticNet
clt=ElasticNet()
clt.fit(X_train,y_train)
y_pred=clt.predict(X_test)
```

```python
print('R-squared: ',r2_score(y_test,y_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```
