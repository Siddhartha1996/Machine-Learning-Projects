# We now use Clustering to group the data

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('K Means')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++')
y_kmeans=kmeans.fit_predict(df)
y_kmeans
```

```python
plt.scatter(df[y_kmeans==0,8],df[y_kmeans==0,7],
s=20,c='blue',label='Cluster 1')
plt.scatter(df[y_kmeans==1,8],df[y_kmeans==1,7],
s=20,c='pink',label='Cluster 2')
plt.scatter(df[y_kmeans==2,8],df[y_kmeans==2,7],
s=20,c='green', label='Cluster 3')
plt.scatter(df[y_kmeans==3,8],df[y_kmeans==3,7],
s=20,c='cyan', label='Cluster 4')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
s=30,c='red',label='centroids')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```

#  Feature Selection 


```python
#Using Lasso CV
df1 = pd.read_csv("CC GENERAL.csv")
df1 = df1.drop('CUST_ID', axis = 1)
df1 = df1.dropna(axis=1)
from sklearn.linear_model import LassoCV

#Feature Selection
X = df1.drop("BALANCE",1)   #Feature Matrix
y = df1["BALANCE"]          #Target Variable

reg = LassoCV()
reg.fit(X, y)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  
      str(sum(coef == 0)) + " variables")
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range (1,30):
    kmeans=KMeans(i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,30),wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
